```python
import matplotlib.pyplot as plt
plt.style.use('mystyle.mplstyle')
```

```python
from matplotlib.lines import Line2D

labels_lens = (
    r"$\xi_{x, 0}$ [$''$]",
    r"$\xi_{y, 0}$ [$''$]",
    r"$\phi_{l}$ [rad]",
    r"$q_{l}$",
    r"$\theta_{E}$ [$''$]",
)
labels_shear = (r"$\gamma_1$", r"$\gamma_2$")
labels_sersic = (
    r"$x_{0}$ [$''$]",
    r"$y_{0}$ [$''$]",
    r"$\phi_{s}$ [rad]",
    r"$q_{s}$",
    r"$n$",
    r"$r_{e}$ [$''$]",
    r"$I_{e}$",
)
labels_sub = (
    r"$\log_{10} m_\mathrm{sub} / \mathrm{M}_\odot$",
    r"$x_\mathrm{sub}$ [$''$]",
    r"$y_\mathrm{sub}$ [$''$]",
)
labels = {
    "z_lens": labels_lens,
    "z_shear": labels_shear,
    "z_sub": labels_sub,
    "z_sersic": labels_sersic,
    "log10_M_smooth": (r"$\log_{10} M_\mathrm{smooth} / \mathrm{M}_\odot$",),
}


fig, axs = plt.subplots(2, 7, figsize=(14, 5.4), dpi = 200)
    
# legend_elements = [Line2D([0], [0], color='tab:orange', lw=2.2, label='Posterior'),
#                    Line2D([0], [0], color='tab:blue', lw=2.2, label='Proposal distribution'),
#                    Line2D([0], [0], color='tab:blue', lw=1.5, ls='--', label=r'New $\Gamma$'),
#                    Line2D([0], [0], color='k', lw=1.5, ls=':', label='True value')
#                   ]

# # Create the figure
# fig.legend(handles=legend_elements, 
#            loc="lower left", bbox_to_anchor=(0.0135, 1.01), 
#            bbox_transform=fig.transFigure, ncol=4,
#           fontsize=12)

for i, ax in zip(range(7), axs[1]):
    ax.tick_params(labelsize=12)
    ax.set_yticks([])
    ax.set_ylabel(None)
    ax.set_xlabel(labels['z_sersic'][i], size=12,labelpad=2)
    
for i, ax in zip(range(7), axs[0]):
    ax.tick_params(labelsize=12)
    ax.set_yticks([])
    ax.set_ylabel(None)
    if i<5:
        ax.set_xlabel(labels['z_lens'][i], size=12, labelpad=2)
    else:
        ax.set_xlabel(labels['z_shear'][i-5], size=12, labelpad=2)
        
axs[0, 0].set_xlim(-0.062, -0.034)
axs[0, 1].set_xlim(-0.005, 0.025)
axs[0, 2].set_xlim(1.48, 1.548)
axs[0, 3].set_xlim(0.495, 0.557)
axs[0, 4].set_xlim(1.611, 1.652)
axs[0, 5].set_xlim(-0.002, 0.02)
axs[0, 6].set_xlim(-0.002, 0.02)

axs[1, 0].set_xlim(-0.013, 0.013)
axs[1, 1].set_xlim(-0.013, 0.013)
axs[1, 2].set_xlim(1.14, 1.26)
axs[1, 3].set_xlim(0.66, 0.73)
axs[1, 4].set_xlim(1.85, 2.14)
axs[1, 5].set_xlim(1.1, 1.5)
axs[1, 6].set_xlim(0.95, 1.4)

fig.patch.set_alpha(0)
for ax in axs.flatten():
    ax.patch.set_alpha(0)

plt.subplots_adjust(wspace=0.18, hspace=0.3)
# plt.tight_layout()
```
